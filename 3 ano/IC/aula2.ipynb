```python
import pennylane as qml
import matplotlib
import numpy as np
```

```python
n_qubits = 2

dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev)
def circuit():
    # qml.PauliX(wires=0)
    qml.Hadamard(wires=[0])
    # qml.Hadamard(wires=[1])
    qml.CNOT(wires=[0, 1])
    

    
    return qml.probs(wires=[0, 1])

print(circuit())
qml.draw_mpl(circuit)()

```
